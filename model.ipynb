## Useful Functions and processing 

```python
def format_dataset(df):
    def transform(x:str):
        index=x.find("-1")
        return x[index+3:]
    data_col = df["packet_dat"]
    new_data_col = data_col.apply(transform)
    df["packet_dat"]=new_data_col

    df['encoded_attack_cat'] = df['attack_cat'].astype('category').cat.codes
    
    return df
```

```python
left_out = full_data.iloc[50000:]
start=50000
stop=50000+50000
for i in range(left_out.shape[0]//50000):
    name = f"batch_{i}.csv"

    batch = left_out.iloc[start:stop]
    batch.to_csv(name,index=False)
    start=stop
    stop+=50000
    print(f"Saving batch {i} of shape :{batch.shape}")
```
